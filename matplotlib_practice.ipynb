```python
# Magic command to display charts
%matplotlib inline
```

```python
# Import dependencies and rename for easier referencing later
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the chart
plt.plot(x_axis,y_axis)
```

```python
# Create line chart using object oriented interface
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, color="g", marker="D", linewidth=2, label="Boston")
ax.set_title("PyBer Fare per Month")
ax.set_xlabel("Months")
ax.set_ylabel("Fare ($)")
ax.set_ylim(0, 45)
ax.grid()
```

```python
# Create plot with broken out fig and ax
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis,y_axis)
```

```python
# Create plot and label for the legend
plt.plot(x_axis, y_axis, marker="*", color="b", linewidth=2, label="Boston")
# Create labels for the X and Y axes
plt.xlabel("Date")
plt.ylabel("Fare ($)")
# Set the Y limit to be within 0 and 45
plt.ylim(0, 45)
# Create a title
plt.title("PyBer Fare Per Month")
# Add a grid
plt.grid()
# Add the legend
plt.legend()
```

```python
# Create a bar plot
plt.bar(x_axis,y_axis)
```

```python
# Creat the bar plot
plt.bar(x_axis, y_axis, color="g", label="Boston")
# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title
plt.title("PyBer Fare by Month")
# Add the legend
plt.legend
```

```python
# Create horizontal bar chart
plt.barh(x_axis, y_axis)
# Use get current axis(gca) to invert
plt.gca().invert_yaxis()
```

```python
# Create plot
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="magenta")
ax.legend("Boston")
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.invert_yaxis()
```

```python
# Create a bar chart using object method
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
# Create horizontal bar chart
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan")
ax.legend("Chicago")
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.invert_yaxis()
```

```python
# Create scatter plot
plt.plot(x_axis, y_axis, 'o')
```

```python
# Create scatter plot with scatter
plt.scatter(y_axis, x_axis, color="r")
plt.legend("Chicago")
plt.title("PyBer fare by Month")
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.xlim(0, 50)
plt.gca().invert_yaxis()
```

```python
# Create the plot
# List comprehension to increase size of dots
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# Create plot using object oriented method
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[i * 5 for i in y_axis], color="skyblue", ec="black", alpha=.2, linewidth=2)
ax.legend("Boston")
ax.set_title("PyBer Fare by Month")
ax.set_xlim(0, 50)
ax.invert_yaxis()
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Assign 12 colors for formatting months
colors = ["slateblue", "lightpink", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0,0,0.2,0,0,0,0.2,0,0,0,0,0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%', colors=colors)

plt.show()
```

```python
# Customizing and creating pie chart with object method
fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(y_axis, labels=x_axis, autopct='%.1f%%', explode=explode_values, startangle=260, shadow=True, colors=colors)
ax.invert_yaxis()

plt.show()
```

```python
# Add magic line
%matplotlib inline
# Import dependencies
import matplotlib.pyplot as plt
import statistics
import numpy as np
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Get standard deviation of the values in the y-axis
stdev = statistics.stdev(y_axis)
stdev
```

```python
# Plot standard deviation
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
# Add error bars to line plot using object method
fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
# Add error bars to bar chart
plt.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
# Add error bars to bar chart using object method
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
# Adjusting ticks
plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0, 51, step=5.0))
plt.gca().invert_yaxis()
from matplotlib.ticker import MultipleLocator
```

```python
# Increase size of figure
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
# Customize ticks using object method
ax.set_xticks(np.arange(0, 51, step=5.0))

# Create minor ticks at an increment of 1
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.show()
```
